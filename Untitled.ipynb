```python
a = int(input("Enter the battery percentage: "))
if a >= 70:
    print("Battery is full")
elif a >= 30:
    print("Current percentage is enough")
else:
    print("Need to charge the mobile")
```

```python
a = input("Enter the color: ")
if a == "red":
    print("STOP")
elif a == "yellow":
    print("Get ready")
else:
    print("Go")
```
